```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('PP_desercion.xlsx')
```

```python
#Eliminar columnas
```

```python
if 'Matricula' in df.columns:
    df = df.drop(columns=['Matricula'])
```

```python
#Renombrar las columnas
```

```python
df = df.rename(columns={
    'Carrera que cursas': 'Carrera',
    'Semestre actual': 'Semestre',
    'Estado civil': 'Estado_Civil',
    '¿Has pensado en abandonar tus estudios?': 'Riesgo_Abandono',
    'Si respondiste “Sí”, ¿Cuál es la principal razón?': 'Razon_Abandono',
    '¿Con qué frecuencia asistes a clases?': 'Frecuencia_Asistencia',
    '¿Qué tan satisfecho(a) estás con la calidad educativa de tu universidad?': 'Satisfaccion_Calidad',
    '¿Consideras que tus profesores te motivan a continuar tus estudios?': 'Motivacion_Prof',
    '¿Trabajas actualmente?': 'Trabaja',
    '¿Aproximadamente cuánto ganas al mes?': 'Ingreso_Propio',
    '¿Tus ingresos o los de tu familia son suficientes para cubrir tus gastos escolares?': 'Suficiencia_Ingreso',
    '¿Aproximadamente cuánto gana al mes tu familia?': 'Ingreso_Familiar',
    '¿En qué gastas principalmente tu dinero cada mes?': 'Gasto_Principal',
    '¿Cuentas con algún tipo de beca o apoyo económico?': 'Tiene_Beca',
    'En caso de contar con algún tipo de beca o apoyo, ¿de cuánto es el monto?': 'Monto_Beca',
    '¿Qué tan importante consideras el factor económico en tu permanencia escolar?': 'Importancia_Economica',
    '¿A qué distancia vives de la universidad?': 'Distancia',
    '¿Cuál es tu medio principal de transporte?': 'Transporte',
    '¿Consideras que la distancia o el transporte influyen en tu asistencia o rendimiento académico?': 'Influencia_Transporte',
    '¿Te sientes apoyado(a) por tu familia o amigos para continuar estudiando?': 'Apoyo_Social',
    '¿Has sentido estrés, ansiedad o desmotivación por cuestiones escolares?': 'Salud_Mental',
    '¿Cómo calificas tu sentido de pertenencia a la universidad?': 'Sentido_Pertenencia',
    'En una escala del 1 al 10, ¿Qué tan comprometido(a) te sientes con concluir tu carrera?': 'Nivel_Compromiso',
    '¿Conoces personas que hayan desertado de esta universidad?': 'Conoce_Desertores',
    '¿Cuales son las razones por las que abandonaron sus estudios?': 'Razon_Desercion_Terceros'
})
```

```python
import matplotlib.pyplot as plt

# PASO 1: Filtrar
# Creamos una lista pequeña solo con los estudiantes que dijeron 'Sí' han pensado abandonar
solo_riesgo = df[df['Riesgo_Abandono'] == 'Sí']

# PASO 2: Contar
# Contamos cuántas veces se repite cada razón
conteo_razones = solo_riesgo['Razon_Principal'].value_counts() # Corrected column name

# PASO 3: Graficar
# 'barh' significa 'bar horizontal'
conteo_razones.plot(kind='barh', color='salmon', figsize=(10, 5))

plt.title('Razón principal de los estudiantes en riesgo')
plt.xlabel('Cantidad de estudiantes')
plt.show()
```

```python
# PASO 1: Cruce de datos
# Cruzamos la columna 'Trabaja' contra 'Riesgo_Abandono'
tabla = pd.crosstab(df['Trabaja'], df['Riesgo_Abandono'])

# PASO 2: Graficar
# stacked=False pone las barras una al lado de la otra
tabla.plot(kind='bar', color=['skyblue', 'orange'], figsize=(8, 6))

plt.title('¿El trabajo influye en el riesgo de abandono?')
plt.ylabel('Cantidad de Estudiantes')
plt.xticks(rotation=0) # Esto es para que el texto de abajo no salga chueco
plt.show()
```

```python
#Salud mental vs riesgo de abandono
```

```python
# Cruzamos 'Salud Mental' vs 'Riesgo de Abandono'
tabla_salud = pd.crosstab(df['Salud_Mental'], df['Riesgo_Abandono'])

# Usamos 'stacked=True' (apilada) para ver qué proporción representa el riesgo
tabla_salud.plot(kind='bar', stacked=True, color=['lightgreen', 'salmon'], figsize=(8, 6))

plt.title('Impacto del Estrés/Ansiedad en la intención de abandonar')
plt.ylabel('Número de Estudiantes')
plt.xticks(rotation=0) # Texto derecho
plt.legend(title='¿Piensa abandonar?')
plt.show()
```

```python
#Satisfacción educativa vs riesgp
```

```python
# Cruzamos qué tan satisfechos están vs si se quieren ir
tabla_calidad = pd.crosstab(df['Satisfaccion_Calidad'], df['Riesgo_Abandono'])

tabla_calidad.plot(kind='bar', color=['skyblue', 'red'], figsize=(10, 6))

plt.title('Relación entre Calidad Educativa y Deserción')
plt.ylabel('Número de Estudiantes')
plt.xticks(rotation=45) # Rotamos el texto un poco para que se lea bien
plt.legend(title='¿Piensa abandonar?')
plt.show()
```

```python
#Convertir el riesgo a 0 y 1
```

```python
# Si dice 'Sí' vale 1 (100%), si dice 'No' vale 0 (0%)
df['Riesgo_Num'] = df['Riesgo_Abandono'].apply(lambda x: 1 if x == 'Sí' else 0)
```

```python
# 1. Probabilidad según si TRABAJA
prob_trabajo = df.groupby('Trabaja')['Riesgo_Num'].mean()
print("--- Probabilidad de Abandono si Trabaja ---")
print(prob_trabajo)
print("-" * 30)

# 2. Probabilidad según su SALUD MENTAL (Estrés/Ansiedad)
prob_salud = df.groupby('Salud_Mental')['Riesgo_Num'].mean()
print("--- Probabilidad de Abandono según Salud Mental ---")
print(prob_salud)
print("-" * 30)

# 3. Probabilidad según APOYO FAMILIAR
prob_apoyo = df.groupby('Apoyo_Social')['Riesgo_Num'].mean()
print("--- Probabilidad de Abandono según Apoyo Familiar ---")
print(prob_apoyo)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculamos la probabilidad (el promedio)
probabilidad = df.groupby('Trabaja')['Riesgo_Num'].mean().reset_index()

probabilidad['Riesgo_Num'] = probabilidad['Riesgo_Num'] * 100

plt.figure(figsize=(8, 6))

# Creamos la gráfica de barras
# 'barplot' calcula el promedio automáticamente y nos dibuja la probabilidad
sns.barplot(x='Trabaja', y='Riesgo_Num', data=probabilidad, palette='coolwarm')

plt.title('Probabilidad de Deserción: ¿Trabajar aumenta el riesgo?')
plt.ylabel('Probabilidad de Abandonar (%)')
plt.xlabel('¿El estudiante trabaja?')

# Ponemos una línea punteada en el 50% como referencia
plt.axhline(50, color='gray', linestyle='--')
plt.text(0.5, 52, 'Línea de riesgo crítico (50%)', color='gray')

plt.show()
```
